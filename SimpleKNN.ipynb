Import necessary modules 

```
from sklearn.neighbors import KNeighborsClassifier 
from sklearn import metrics
from sklearn.model_selection import train_test_split 
from sklearn.datasets import load_iris 
import numpy as np 
import matplotlib.pyplot as plt 
```

```
#iris dataset
irisData = load_iris() 

X = irisData.data 
Y = irisData.target 


#Here, train_test_split function splits data into two subsets(training data and testing data)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42) 
#test_size parameter is passed to train_test_split function to decide the size of the data that has to be split as the test dataset

neighbors = np.arange(1, 15)
#Create empty arrays
train_accuracy = np.empty(len(neighbors)) 
test_accuracy = np.empty(len(neighbors)) 


#Loop : k nearest neighbors
for i, k in enumerate(neighbors): 
	knn = KNeighborsClassifier(n_neighbors=k) 
	knn.fit(X_train, Y_train) 
	
	# Compute traning and test data accuracy 
	train_accuracy[i] = knn.score(X_train, Y_train) 
	test_accuracy[i] = knn.score(X_test, Y_test) 
```

Plot

```
plt.figure(figsize=(10,8))
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy') 
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy') 
plt.title('Accuracy plot')
plt.legend() 
plt.xlabel('Neighbors') 
plt.ylabel('Accuracy') 
plt.show() 
```
